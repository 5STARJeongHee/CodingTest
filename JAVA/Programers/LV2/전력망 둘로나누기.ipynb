```html
n개의 송전탑이 전선을 통해 하나의 트리 형태로 연결되어 있습니다. 당신은 이 전선들 중 하나를 끊어서 현재의 전력망 네트워크를 2개로 분할하려고 합니다. 이때, 두 전력망이 갖게 되는 송전탑의 개수를 최대한 비슷하게 맞추고자 합니다.

송전탑의 개수 n, 그리고 전선 정보 wires가 매개변수로 주어집니다. 전선들 중 하나를 끊어서 송전탑 개수가 가능한 비슷하도록 두 전력망으로 나누었을 때, 두 전력망이 가지고 있는 송전탑 개수의 차이(절대값)를 return 하도록 solution 함수를 완성해주세요.
```

````java
import java.util.*;

HashMap<Integer, Integer> parent = new HashMap<>();
HashMap<Integer, Integer> rank = new HashMap<>();
public int solution(int n, int[][] wires) {
        int answer = 1000;
        
        for(int i = 0; i<wires.length; i++){
            int[] save = wires[i].clone();
            wires[i][0] = 0;
            wires[i][1] = 0;
            
            ArrayList<Integer> result = makeGraph(n, wires);
        
            System.out.println(result.get(0) + ","+result.get(1));
            int diff = Math.abs(result.get(0) - result.get(1));
            wires[i][0] = save[0];
            wires[i][1] = save[1];
            if(diff < answer) {
                answer = diff;
            }
            // System.out.println("answer = " +answer);
        }
        
    
        return answer;
}

private void makeSet(Integer node) {
    parent.put(node, node);
    rank.put(node, 0);
}
private Integer find(Integer node){
    if(parent.get(node) != node){
        parent.put(node, find(parent.get(node)));
    }
    return parent.get(node);
}
private void union(Integer node1, Integer node2){
    Integer root1 = find(node1);
    Integer root2 = find(node2);
    if(rank.get(root1) > rank.get(root2)){
        parent.put(root1, root2);
    }else {
        parent.put(root1, root2);
        if(rank.get(root1) == rank.get(root2)){
            rank.put(root2, rank.get(root2) + 1);
        }
    }
}

private ArrayList<Integer> makeGraph(int v, int[][] wires){
    for(int i = 1; i<=v; i++ ){
        makeSet(i);
    }

    for(int i = 0; i< wires.length; i++){
        int[] wire = wires[i];
        if(find(wire[0]) != find(wire[1])){
            union(wire[0], wire[1]);
        }
    }
    for(int i = 1 ; i<= v; i++){
        // System.out.println(i +" : " + find(i));
    }

    HashMap<Integer,Integer> edgeCntMap = new HashMap<Integer,Integer>();
    
    for(int i = 1; i<= v; i++){
        Integer p = find(i);
        // System.out.println("parent: " + p);
        Integer cnt = edgeCntMap.get(p);
        if(cnt != null){
            edgeCntMap.put(p, cnt+1);
        } else{
            edgeCntMap.put(p, 1);
        }
    }
    ArrayList<Integer> edgeCnt = new ArrayList<Integer>();

    Iterator<Integer> cntIt = edgeCntMap.values().iterator();
    while(cntIt.hasNext()){
        edgeCnt.add(cntIt.next());
    }
    return edgeCnt;
}
````

## 풀이법 : 

전체 간선에서 하나씩 뺀 간선 목록을 최소신장 트리를 구하는 크루스칼 알고리즘에 적용하여

각 간선이 없는 간선 목록들의 부모노드를 카운트하여 각각 연결된 그래프와 각 간선 수를 세었다.

각 간선 수를 비교하여 최소값을 구한다.

알고리즘에 익숙하지 않아서인지 코드량과 시간에 비하면 생각보다 효율적이지 않은 방법인 것 같다.

```java
import java.util.*;
class Solution {
     // 1. 각 간선 끊기
        // 2. bfs 돌면서 몇개인지 체크
        public int solution(int n, int[][] wires) {
            int answer = Integer.MAX_VALUE;

            ArrayList<Node> arr[] = new ArrayList[n+1];
            for(int i=0;i<n+1;i++){
                arr[i]=new ArrayList<>();
            }

            for(int i=0;i<n-1;i++){
                arr[wires[i][0]].add(new Node(wires[i][1],i));
                arr[wires[i][1]].add(new Node(wires[i][0],i));
            }

            for(int i=0;i<n-1;i++){
                //i번 간선 끊기

                boolean visited[] = new boolean[n+1];

                Queue<Integer> q =new LinkedList<>();
                q.add(1);
                visited[1]=true;
                int cnt=1;
                while(!q.isEmpty()){
                    int nowIndex = q.poll();

                    for(Node nextNode : arr[nowIndex]){
                        if(nextNode.num==i || visited[nextNode.index]){
                            continue;
                        }

                        q.add(nextNode.index);
                        visited[nextNode.index]=true;
                        cnt++;

                    }

                }



                answer=Integer.min(answer,Math.abs(n-cnt-cnt));


            }


            return answer;
        }

        public class Node{
            int index;
            int num;

            public Node(int index, int num) {
                this.index = index;
                this.num = num;
            }
        }
}
```

```java
class Solution {
    int N, min;
    int[][] map;
    int[] vst;
    int dfs(int n){
        vst[n] = 1;
        int child = 1;
        for(int i = 1; i <= N; i++) {
            if(vst[i] == 0 && map[n][i] == 1) {
                vst[i] = 1;
                child += dfs(i);
            }
        }
        min = Math.min(min, Math.abs(child - (N - child)));
        return child;
    }
    public int solution(int n, int[][] wires) {
        N = n;
        min = n;
        map = new int[n+1][n+1];
        vst = new int[n+1];
        for(int[] wire : wires) {
            int a = wire[0], b = wire[1];
            map[a][b] = map[b][a] = 1;
        }
        dfs(1);
        return min;
    }
}
```

## 해석

문제의 핵심은 그래프를 둘로 끊는 것이며 
간선들의 연결된 상태는 둘로 끊긴 그래프 중 하나만 알면 다른 그래프의 간선의 개수는 쉽게 구할 수 있다.

따라서 dfs,bfs 알고리즘을 통해서 연결된 노드들을 모두 방문하게 되면 둘로 끊긴 그래프 중 하나만 전부 방문하게 되고 
방문마다 간선 개수를 세어두면 쉽게 해결이 가능하다.
