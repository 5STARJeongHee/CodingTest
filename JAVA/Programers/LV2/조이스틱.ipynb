조이스틱으로 알파벳 이름을 완성하세요. 맨 처음엔 A로만 이루어져 있습니다.
ex) 완성해야 하는 이름이 세 글자면 AAA, 네 글자면 AAAA

조이스틱을 각 방향으로 움직이면 아래와 같습니다.

▲ - 다음 알파벳
▼ - 이전 알파벳 (A에서 아래쪽으로 이동하면 Z로)
◀ - 커서를 왼쪽으로 이동 (첫 번째 위치에서 왼쪽으로 이동하면 마지막 문자에 커서)
▶ - 커서를 오른쪽으로 이동 (마지막 위치에서 오른쪽으로 이동하면 첫 번째 문자에 커서)
예를 들어 아래의 방법으로 "JAZ"를 만들 수 있습니다.

- 첫 번째 위치에서 조이스틱을 위로 9번 조작하여 J를 완성합니다.
- 조이스틱을 왼쪽으로 1번 조작하여 커서를 마지막 문자 위치로 이동시킵니다.

````java
public int solution(String name) {
    int answer = 0;
    HashMap<Character, Integer> alphaMap = new HashMap<>(){{
        put('A', 0);
        put('B', 1);
        put('C', 2);
        put('D', 3);
        put('E', 4);
        put('F', 5);
        put('G', 6);
        put('H', 7);
        put('I', 8);
        put('J', 9);
        put('K', 10);
        put('L', 11);
        put('M', 12);
        put('N', 13);
        put('O', 14);
        put('P', 15);
        put('Q', 16);
        put('R', 17);
        put('S', 18);
        put('T', 19);
        put('U', 20);
        put('V', 21);
        put('W', 22);
        put('X', 23);
        put('Y', 24);
        put('Z', 25);
    }};
    
    int[] alpha = new int[name.length()];
    boolean[] checked = new boolean[name.length()];

    int next = 0;
    int mv = 0;

    for(int i = 0; i<name.length(); i++){
        if(name.charAt(i) == 'A') {
            checked[i] = true;
        }
    }
    checked[0] = false;
    
    while(checked[next] == false){
        System.out.println("current : " + name.charAt(next) + ", next:" + next);

        char c = name.charAt(next);
        int up = alphaMap.get(c);
        int down = 26-up;

        if(up != 0){
            alpha[next] = Math.min(up,down);
        }
        System.out.println("up, down minimum: " + alpha[next]);
        checked[next] = true;
        int idx = 0;
        int min_dist = Integer.MAX_VALUE;
        for(int i = 1; i<checked.length; i++){
            if(!checked[i]){
                int distance = Math.min(i-next, (next-0) + (checked.length - i));
                if(min_dist > distance){
                    System.out.println("temp distance: " + distance + " idx: " + i);
                    min_dist = distance;
                    idx = i;
                }
            }
        }
        if(idx !=0 ){
            System.out.println("next: " + name.charAt(idx) + ", mv: " + min_dist);
            mv += min_dist;
            next = idx;        
        }
    
    }

    for(int i = 0; i< alpha.length; i++ ){
        answer += alpha[i];
    }
    answer += mv;

    return answer;
}
````

````java
class Solution {
    public int solution(String name) {
        int answer = 0, rem = 0, cnt = Integer.MAX_VALUE;
        // rem 이전 인덱스 정보
        // cnt 이동 거리
        for (int i = 0; i < name.length(); i++) {
            if (name.charAt(i) != 'A') {
                answer += calc(name.charAt(i));

                if (i == 0) continue;
                int tmp = rem + name.length()-i; // 이전 인덱스에서 현재 인덱스까지 왼쪽으로 간 거리
                //rem*2 + name.length()-i , rem + (name.length-i)*2
                cnt = Math.min(cnt, Math.min(tmp+rem, tmp+name.length()-i)); // 현재 최소 이동 거리, (왼쪽 이동 거리에서 이전 인덱스로 이동하는 거리 비교(좌,우))
                System.out.println("now loc: " +i + ", tmp: " + tmp + ", rem: " + rem + ", cnt: " + cnt);
                System.out.println("now answer: " +answer);
                rem = i;
            }
        }
        cnt = Math.min(rem, cnt);
        System.out.println("now cnt:" + cnt);
        return answer+cnt;
    }

    private int calc(char c) {
        int cint = (int) c;
        return Math.min(91-c, c-65);
    }

    public int solution2(String name) {
        int answer = 0, rem = 0, cnt = Integer.MAX_VALUE;
        // rem 이전 인덱스 정보
        // cnt 이동 거리
        for (int i = 0; i < name.length(); i++) {
            if (name.charAt(i) != 'A') {
                answer += calc(name.charAt(i));

                if (i == 0) continue;
                int left = rem + name.length()-i; // 이전 인덱스에서 현재 인덱스까지 왼쪽으로 간 거리
                int right = i - rem; // 오른쪽으로 간 거리
                cnt = Math.min(cnt, Math.min(rem + left, rem + right)); // 현재 최소 이동 거리, (왼쪽 이동 거리에서 이전 인덱스로 이동하는 거리 비교(좌,우))
                
                rem = i;
            }
        }
        cnt = Math.min(rem, cnt);
        return answer+cnt;
    }
}
````
