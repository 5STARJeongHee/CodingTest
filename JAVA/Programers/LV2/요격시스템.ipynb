`````java
import java.util.*;

public int solution(int[][] targets) {
    int answer = 0;
    int min = Integer.MAX_VALUE;
    int max = -1;

    for(int i = 0; i<targets.length; i++){
        min = Math.min(min, targets[i][0]);
        max = Math.max(max, targets[i][1]);
    }

    int d = (max - min -1) * 2;
    ArrayList<ArrayList<Integer>> list = new ArrayList<>();
    int[] numArr = new int[d];
    boolean[] visited = new boolean[d];

    for(int i = 1; i<=d; i++){
        numArr[i-1] = i;
        ArrayList<Integer> missaile = new ArrayList<>();
        for(int j = 0; j<targets.length; j++){
            Double fire = min + 0.5 * i ;
            
            if(fire >targets[j][0] && fire < targets[j][1]){
                missaile.add(j);
            }
        }
        list.add(missaile);
    }

    for(int i = 1; i<d; i++){
        ArrayList<ArrayList<Integer>> combs = new ArrayList<>();
        dfs_combination(visited, numArr, 0, d, i, combs);
        HashSet<Integer> misSet = new HashSet<>();
        for(ArrayList<Integer> t: combs){
            for(Integer k: t){
                misSet.add(k);              
            }
            if(misSet.size() == targets.length){
                return i;
            }
            misSet.clear();
        }
        
    }

    
    return answer;
}
public void dfs_combination(boolean[] visited, int[] arr, int start, int n, int r, ArrayList<ArrayList<Integer>> combs){
    if(r == 0){
        ArrayList<Integer> comb = new ArrayList<>();
        for(int i = 0; i<visited.length; i++){
            if(visited[i]){
                System.out.println(comb.toString());
                comb.add(i);
            }
        }
        combs.add(comb);
    }

    for(int i = start; i<n ; i++){
        visited[i] = true;
        dfs_combination(visited, arr, i+1, n, r-1, combs);
        visited[i] = false;
    }
}
`````

## 조합 계속 까먹어서  정리
조합은 다음과 같이 표현 가능

nCr = n-1Cr-1 + n-1Cr

예를 들어 (1, 2, 3) 중에서 2개를 뽑는 조합이라고 생각해보면 -> 3C2

(1, X) - 1을 뽑는 경우(하나의 원소를 선택할 경우)
(X, X) - 1을 뽑지 않는 경우(하나의 원소를 선택하지 않는 경우) 
이처럼 2가지로 나뉠 수 있다.

 

1을 뽑은 경우 나머지 (2, 3) 중 1개를 선택해야 함. (총 2개의 경우 (1, 2) (1, 3)) -> 2C1
1을 뽑지 않은 경우 (2, 3) 모두 선택해야 한다. (총 1개의 경우 (2, 3)) -> 2C2

3C2 = 2C1 + 2C2 >... > 3C0 = 1이 될 때까지 재귀 호출을 통해 구현한다.

경우의 수만 구한다면 메서드 토대는 다음 처럼 생각해 볼 수 있다

```java
public static int combination(int n, int r) {
    if(n == r || r == 0) 
        return 1; 
    else 
        return combination(n - 1, r - 1) + combination(n - 1, r); 
}
```

경우의 수가 아닌 실제 조합을 뽑아 본다면 다음과 같이 구해볼 수 잇다.

public static void comb1(int[] arr, boolean[] visited, int start, int n, int r) {
	if(r == 0) {
		print(arr, visited, n);
		return;
	} else {
		for(int i = start; i < n; i++) {
			visited[i] = true; // i 를 뽑은 경우

		}
	}
}

            visited[depth] = true; // depth 를 뽑은 경우
            comb2(arr, visited, depth + 1, n, r - 1);

    }
