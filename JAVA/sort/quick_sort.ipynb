# Quick Sort

## - 알고리즘
- Divide and Conquer 를 이용한 정렬방법
- n개의 데이터를 정렬 시, 최악의 경우 O(n^2) 이 걸리며, 일반적으로는 O(nlogn)이 걸린다.
- 재귀적 호출로 인한 O(logn) 만큼의 메모리가 필요, 최악의 경우 O(n) 

## 정렬

1. 리스트내의 원소들 중 하나를 선택 -> 피벗이라고 부르며 보통 맨 왼쪽, 중간, 맨 오른쪽 원소를 선택한다.
2. 순서대로 원소들을 피벗과 비교하여 피벗보다 작은 원소들을 먼저 비교한 순으로 피벗 왼쪽에 둔다(자연스럽게 피벗 오른쪽은 피벗보다 큰 수가 된다.)
3. 비교가 끝나고나면 피벗보다 작은 수는 왼쪽, 큰수는 오른쪽에 있지만 피벗은 자리를 바꾸지 않았으므로 중간에 들어가도록 이동
4. 피벗보다 작은 수들로 이뤄진 리스트와 큰 수들로 이뤄진 리스트로 나누어 각각 1~3을 진행
5. 재귀가 끝나고 나면 자동 정렬되어 있다.

출처: 위키

```python
import java.util.*;
public class QuickSort{
    public List<Integer> quickSort(List<Integer> arr, int start, int end){
        // pivot 
        if (start == end) return arr;
        int p = arr.get(end);
        int index = start - 1;
        // 피벗 보다 작은 수를 피벗 보다 왼쪽에 정렬
        System.out.println("p: " +p);
        for(int j = 0; j<arr.size(); j++){
            if(arr.get(j) < p){
                index++;
                System.out.println(index + ", " + j + " value : " + arr.get(index) + ": " + arr.get(j));
                Collections.swap(arr, index, j);
            }
        }
        // 아직 정렬되지 않은 피벗 정렬 -> index - 피벗 위치가 됨
        Collections.swap(arr, index++, end);
        // 피벗 좌우로 divide
        quickSort(arr, start, index-1);
        quickSort(arr, index+1, end);
        return arr;
    }
}
```

```python
ArrayList<Integer> testData = new ArrayList<Integer>();

for (int index = 0; index < 100; index++) {
    testData.add((int)(Math.random() * 100));
}
QuickSort sort = new QuickSort();

sort.quickSort(testData,0, testData.size()-1);
```
